```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
raiden = cv2.imread("raiden.jpg")
plt.imshow(mpimg.imread("raiden.jpg"))
print(raiden.shape)
```

```python
hehe = cv2.imread("hehe.jpg")
plt.imshow(mpimg.imread("hehe.jpg"))
print(hehe.shape)
```

```python
# mengubah ukuran citra untuk menyamakan ukuran
hehe = cv2.resize(hehe, (raiden.shape[0], raiden.shape[1]))
hehe.shape
```

```python
raiden_hehe = cv2.add(raiden, hehe)
cv2.imwrite("raiden_hehe.jpg", raiden_hehe)
plt.imshow(mpimg.imread("raiden_hehe.jpg"))
```

```python
# gabung vertikal
vertikal_raiden_hehe = np.vstack((raiden, hehe))
cv2.imwrite("vertikal_raiden_hehe.jpg", vertikal_raiden_hehe)
plt.imshow(mpimg.imread("vertikal_raiden_hehe.jpg"))
```

```python
# gabung horizontal
horizontal_raiden_hehe = np.hstack((raiden, hehe))
cv2.imwrite("horizontal_raiden_hehe.jpg", horizontal_raiden_hehe)
plt.imshow(mpimg.imread("horizontal_raiden_hehe.jpg"))
```

```python
# posisi citra kedua di citra pertama
x = 3
y = 5
width = 75
height = 85

raiden_hehe2 = raiden.copy()
raiden_hehe2[y:y+height, x:x+width] = cv2.resize(hehe, (width, height))
cv2.imwrite('raiden_hehe2.jpg', raiden_hehe2)
plt.imshow(mpimg.imread("raiden_hehe2.jpg"))
```
